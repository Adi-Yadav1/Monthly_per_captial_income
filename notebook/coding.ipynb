```python
# Pie chart of average expenditure composition

# Data
expenditure_cols = [
    'Food_Expenditure',
    'Housing_Expenditure',
    'Transport_Expenditure',
    'Healthcare_Expenditure',
    'Other_Expenditures'
]
avg_expense = df[expenditure_cols].mean()

# Color palette
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']

# Plot
plt.figure(figsize=(9, 9))
wedges, texts, autotexts = plt.pie(
    avg_expense,
    labels=avg_expense.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops={'fontsize': 12, 'color': 'black'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)

# Style tweaks
plt.setp(autotexts, size=13, weight="bold")
plt.title('Average Household Expenditure Composition', fontsize=16, weight='bold', color='#333')
plt.axis('equal')  # Keeps the pie as a circle

# Optional: Add legend if you want more clarity
plt.legend(wedges, avg_expense.index, title="Expenditure Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.set(style='whitegrid')

# Color palette
palette = sns.color_palette("coolwarm", len(df['Education_Level'].unique()))

# Create the countplot
sns.countplot(
    x='Education_Level',
    data=df,
    order=df['Education_Level'].value_counts().index,
    palette=palette,
    edgecolor='black'
)

# Title and labels
plt.title('Distribution of Education Levels Among Household Heads', fontsize=16, weight='bold', color='#333')
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Number of Households', fontsize=12)

# Aesthetic tweaks
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Annotate bars with counts
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', fontsize=10, color='black', xytext=(0, 8),
                       textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Get the counts of vehicle ownership categories
vehicle_counts = df["Vehicle_Ownership"].value_counts()

# Define labels and colors for the pie chart
vehicle_labels = vehicle_counts.index.astype(str).tolist()  # dynamic label generation
vehicle_colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # Choosing a palette with good contrast

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(vehicle_counts, labels=vehicle_labels, autopct='%1.1f%%', colors=vehicle_colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Set the title and improve readability
plt.title("Households with Vehicle Ownership", fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.

# Display the plot
plt.show()
```

```python
# Automatically select relevant expenditure columns based on column names
expense_categories = [col for col in df.columns if "Expenditure" in col or "Expense" in col]

# Check if any expenditure-related columns exist in the dataset
if not expense_categories:
    print("Error: No expenditure-related columns found. Please check your dataset.")
else:
    # Sum the total expenses for each category
    expense_sums = df[expense_categories].sum()

    # Create the pie chart with enhanced visuals
    plt.figure(figsize=(8, 8))

    # Define a more aesthetically pleasing color palette
    pie_colors = plt.cm.Paired(range(len(expense_categories)))

    # Plot the pie chart with labels and percentage display
    plt.pie(expense_sums, labels=expense_categories, autopct='%1.1f%%', 
            colors=pie_colors, startangle=140, wedgeprops={'edgecolor': 'black'})

    # Set the title and customize the font size
    plt.title("MPCE Breakdown by Expense Type", fontsize=16, fontweight='bold')

    # Ensure the pie chart is circular and tidy
    plt.axis('equal')

    # Display the chart
    plt.tight_layout()
    plt.show()
```
